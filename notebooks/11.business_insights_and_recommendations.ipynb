# Pesa Salama: Business Insights and Strategic Recommendations

This notebook analyzes sentiment patterns, fraud indicators, customer distress trends, and operational risks within Kenya's fintech ecosystem to generate actionable business insights and recommendations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('pesa_salama_MASTER_tableau.csv')

df.head()
```

```python
df.columns
```

# Business Insights Summary

This section evaluates major customer complaint patterns, fraud indicators, sentiment behavior, and financial distress trends across fintech applications.

The objective is to transform analytical findings into actionable business recommendations that can improve customer experience, reduce fraud risks, and strengthen fintech operational performance.

# Customer Sentiment Trend Analysis

Customer sentiment analysis helps identify overall public perception toward fintech platforms. Negative sentiment patterns may indicate dissatisfaction related to transaction failures, delayed customer support, fraud concerns, or poor application performance.

```python
df['sentiment'].value_counts()
```

### Business Insight

The analysis shows that negative customer sentiment forms a significant portion of fintech reviews. This suggests that many users experience operational frustrations, failed transactions, fraud-related concerns, and dissatisfaction with service delivery.

Persistent negative sentiment can damage customer trust, reduce retention, and negatively affect platform reputation within Kenya’s competitive fintech ecosystem.

```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x='sentiment',
    order=df['sentiment'].value_counts().index
)

plt.title('Customer Sentiment Distribution')

plt.show()
```

# Fraud Complaint Analysis

Fraud-related complaints represent one of the most critical operational risks within digital financial platforms. Identifying fraud patterns helps fintech companies improve transaction monitoring, customer protection, and security response mechanisms.

```python
df.groupby('app_name')['fraud_flag'].mean().sort_values(ascending=False)
```

### Business Insight

Some fintech applications demonstrate higher fraud complaint rates than others, indicating possible weaknesses in fraud prevention systems, transaction verification procedures, or customer dispute resolution processes.

Applications with elevated fraud indicators may experience reduced consumer confidence and increased regulatory scrutiny if fraud mitigation measures are not strengthened.

```python
fraud_rates = (
    df.groupby('app_name')['fraud_flag']
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

fraud_rates.plot(kind='bar')

plt.title('Fraud Complaint Rate by Application')

plt.ylabel('Fraud Rate')

plt.show()
```

# Financial Distress Score Analysis

The financial distress score combines fraud complaint rates and negative sentiment indicators to estimate customer dissatisfaction and operational risk levels within fintech applications.

Higher distress scores may signal increased customer frustration, reputational risk, and declining service quality.

```python
distress = (
    df.groupby('app_name')['distress_score']
    .mean()
    .sort_values(ascending=False)
)

distress
```

### Business Insight

Applications with higher financial distress scores are more likely to experience operational instability, customer dissatisfaction, and trust-related challenges.

These findings suggest that customer sentiment and fraud concerns are closely linked to broader perceptions of fintech reliability and financial safety.

```python
plt.figure(figsize=(8,5))

distress.plot(kind='bar')

plt.title('Average Financial Distress Score by Application')

plt.ylabel('Distress Score')

plt.show()
```
