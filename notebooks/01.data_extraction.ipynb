## Data Collection

### 1.1 Data Source

**Source:** Google Play Store — Kenya country store (`country='ke'`)

**Method:** Scraped using the open-source `google-play-scraper` Python library. No API key or login required.

**Why this source:**
- Raw, unprocessed, user-generated text — no curation applied
- Built-in ground-truth labels (star ratings) for supervised learning
- Multilingual: English, Swahili, and Sheng
- Kenya-specific store ensures data is from Kenyan users

**Apps to be scraped:**

| App | Play Store ID | Role |
|-----|--------------|------|
| M-PESA | `com.safaricom.mpesa.lifestyle` | Core mobile money transactions |
| MySafaricom | `com.safaricom.mysafaricom` | Safaricom account management |
| KCB Mobile | `com.kcb.mobilebanking.android.mbp` | Traditional bank digital app |
| Equity Mobile | `ke.co.equitygroup.equitymobile` | Kenya's largest bank by customers |
| Tala | `com.inventureaccess.safarirahisi` | Digital micro-lending |
| Branch | `com.branch_international.branch.branch_demo_android` | Digital micro-lending |

---



### 1.2 Import libraries

```python
#Import Libraries
from google_play_scraper import reviews, Sort
import pandas as pd
import time
```

### 1.3 Scrape the Apps

We scrape all apps in one code and save to one CSV immediately after collection.

> **⏱️ Expected time:** 2 minutes for all 6 apps
We concanate all 6 app  into one single master CSV.

**Important:** This raw file is **never modified** after saving. It is the permanent, untouched record of all data exactly as collected. All cleaning happens later in Section 3 (Preprocessing).

```python
# Define the apps to scrape
APPS = {
    "mpesa":       "com.safaricom.mpesa.lifestyle",      
    "mysafaricom": "com.safaricom.mysafaricom",           
    "kcb":         "com.kcb.mobilebanking.android.mbp",   
    "equity":      "ke.co.equitygroup.equitymobile",      
    "tala":        "com.inventureaccess.safarirahisi",    
    "branch":      "com.branch_international.branch.branch_demo_android", 
}

all_dfs = []

for app_name, app_id in APPS.items():
    print(f"Scraping {app_name} — ID: {app_id}")
    try:
        result, _ = reviews(
            app_id,
            lang    = 'en',
            country = 'ke',
            sort    = Sort.NEWEST,
            count   = 10000,
        )
        df = pd.DataFrame(result)
        df['app_name'] = app_name
        all_dfs.append(df)
        print(f"  ✓ {len(df):,} rows")
    except Exception as e:
        print(f"  ✗ Failed: {e}")
    time.sleep(3)
# save master file to csv
master = pd.concat(all_dfs, ignore_index=True)
master.to_csv('master_raw_kenya_fintech.csv', index=False, encoding='utf-8-sig')
```
